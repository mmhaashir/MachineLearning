# Categorical Plots

**Statistical Estimation Within Categories**

The categorical plots display statistical metrics per category. For example, mean value per category or the count of the number of rows per category.

The two main types of plot for this are:

* countplot(): Counts the number of rows per category.
* barplot(): General form of displaying any chosen metric per category.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/kaggle/input/dm-sales/dm_office_sales.csv')

df.head()
```

```python
# Counting the unique values for division
df['division'].value_counts()
```

**Count Plot**

```python
# Creating a count plot for the division
plt.figure(figsize=(10,4))
sns.countplot(x='division', data=df)
```

```python
df['level of education'].value_counts()
```

```python
plt.figure(figsize=(10,4))
sns.countplot(x='level of education', data=df)
```

```python
# We can add a hue parameter to get more insights
plt.figure(figsize=(10,4))
sns.countplot(x='level of education', data=df, hue='training level')
```

**Distribution Within Categories**

We'll discuss the following types:

* Boxplot
* Violinplot
* Swarmplot
* Boxenplot

```python
df = pd.read_csv('/kaggle/input/student-performance/StudentsPerformance.csv')
df.head()
```

**Boxplot**

A boxplot is a graph that gives a visual indication of how a data set's mean, median, mode, minimum, maximum and outlier values are spread out and compare to each other. Boxplots are drawn as a box with a vertical line down the middle, and has horizontal lines attached to each side (known as “whiskers”).

```python
# Creating a boxplot

# It shows the distribution of the data. The edges of the blue box shows the Q1, and Q3,
# the line in the middle of the box shows the median, the whisker above shows the max,
# the whisker below shows the min, and all the data out of the whiskers are the outliers

sns.boxplot(data=df, y='math score')
```

```python
# We can compare the distribution of two column
sns.boxplot(x='math score', y='test preparation course', data=df)
```

**Violinplot**

A violin plot is a hybrid of a box plot and a kernel density plot, which shows peaks in the data. It is used to visualize the distribution of numerical data. Unlike a box plot that can only show summary statistics, violin plots depict summary statistics and the density of each variable.

```python
# Creating a violinplot

sns.violinplot(data=df, x='reading score', y='parental level of education')

# Notice, the plots are a bit violent which shows the KDE
```

**Swarmplot**

A swarm plot is a type of scatter plot that is used for representing categorical values. It is very similar to the strip plot, but it avoids the overlapping of points. We can use the seaborn. swarmplot() to create such graphs. It is not advisable to use this type of graph when the sample size is large.

```python
sns.swarmplot(data=df, x='math score')
```

```python
plt.figure(figsize=(10,4), dpi=200)
sns.swarmplot(data=df, x='parental level of education', y='math score', hue='gender')
```

**Boxenplot**

It is similar to a box plot in plotting a nonparametric representation of a distribution in which all features correspond to actual observations. By plotting more quantiles, it provides more information about the shape of the distribution, particularly in the tails.

```python
sns.boxenplot(data=df, x='math score', y='parental level of education', hue='gender')
```
