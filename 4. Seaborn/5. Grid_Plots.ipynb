# Grid Plots

Seaborn grids calls use Matplotlib subplots() to automatically create a grid based off a categorical column.

Instead of passing in a specific number of columns or rows for the subplots, we can simply pass in the name of the column and seaborn will automatically map subplot grid.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Reading the dataset
df = pd.read_csv('/kaggle/input/student-performance/StudentsPerformance.csv')
df.head()
```

**Catplot**

Categorical plots show the relationship between a numerical and one or more categorical variables.

```python
# Creating a catplot
sns.catplot(data=df, x='gender', y='math score', kind='box')
```

```python
# We can add some other parameters like row and column, which basically works the same as
# hue, but instead of different coloring as in hue, it will create different plots for each
# entry in the column specified

# Using the row parameter
sns.catplot(data=df, x='gender', y='math score', kind='box', row='lunch')
```

```python
# Using the col parameter
sns.catplot(data=df, x='gender', y='math score', kind='box', col='lunch')
```

```python
# Using both
sns.catplot(data=df, x='gender', y='math score', kind='box', col='lunch', row='test preparation course')
```

**Pairgrid**

```python
# Creating a pairgrid
sns.PairGrid(data=df)

# As seen in the output, it gives out empty plots. We need to tell it what to fill in in
# these plots
```

```python
g = sns.PairGrid(data=df)

# We are telling it to fill the upper diagonal with scatterplots
g = g.map_upper(sns.scatterplot)

# We are telling it to fill the diagonal with histograms
g = g.map_diag(sns.histplot)

# We are telling it to fill the lower diagonal with kde plots
g = g.map_lower(sns.kdeplot)
```
