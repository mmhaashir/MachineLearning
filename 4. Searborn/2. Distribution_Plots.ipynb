# Distribution Plots

Distribution plots displays a single continuous feature and help visualize properties such as deviation and average values

There are 3 main distribution plot types;

* Rug Plot
* Histogram
* KDE Plot

**Rug Plot**

A rug plot is the simplest distribution plot and merely adds a dash or tick line for every single value.

The y-axis does not really have a meaning in it.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/kaggle/input/dm-sales/dm_office_sales.csv')
df.head()
```

```python
# As we do not have a meaningful y-axis, we just pass the x-axis

sns.rugplot(x='salary', data=df)
```

```python
# We can add height to the plot. We can do this by passing a value to the "height" param.
# The height param takes values between 0 and 1. So 0.5 would mean to take half the height
# of the y-axis.

sns.rugplot(x='salary', data=df, height=0.5)
```

**Histogram**

If we count how many ticks there are per various x-ranges, we can create a "histogram".

A histogram is a graphical representation of data points organized into user-specified ranges. Similar in appearance to a bar graph, the histogram condenses a data series into an easily interpreted visual by taking many data points and grouping them into logical ranges or bins.

```python
# There are two common ways of achieving it.

# Method 1
sns.displot(x='salary', data=df)
```

```python
# Method 2

sns.histplot(x='salary', data=df)
```

```python
# We can get multiple bins. Let's divide into 50 bins
sns.histplot(x='salary', data=df, bins=50)
```

```python
# We can perform some styling too.

sns.histplot(x='salary', data=df, color='red', edgecolor='blue', linewidth = 2)
```

**KDE**

Kernel density estimation is the process of estimating an unknown probability density function using a kernel function . While a histogram counts the number of data points in somewhat arbitrary regions, a kernel density estimate is a function defined as the sum of a kernel function on every data point.

```python
# We can get this my adding "kde=True"
sns.histplot(x='salary', data=df, kde=True)
```

```python
# We can get all three of them at the same time. Remember, it can be done only with
# displot

sns.displot(x='salary', data=df, kde=True, rug=True)
```
