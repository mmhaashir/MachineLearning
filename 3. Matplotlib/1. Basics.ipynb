# Matplotlib

Matplotlib is sometimes known as the "grandfather" of visualization. Many other visualization libraries are directly built on Matplotlib which includes the famous library "Seaborn"

Matplotlib is heavily inspired by the plotting functions of Matlab. It allows for the creation of almost any type of plot and heavy customization

**A Simple Plot**

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Creating a simple plot

x = np.arange(0,10)
y = 2*x

# Plot is a function of pyplot, which accepts x and y coordinates and plot them
plt.plot(x,y)

# This adds a title to the graph
plt.title("A linear graph")

# This function adds x axis label
plt.xlabel("X-Axis")

# This function adds y axis label
plt.ylabel("Y-Axis")


# This function displays the plot
plt.show()

# This functions saves the plot image
plt.savefig('my_first_plot')
```
