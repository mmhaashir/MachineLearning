# Some Useful Methods

**The Apply Method**

```python
df = pd.read_csv('/kaggle/input/tipscsv/tips.csv')

# The apply method lets you apply any custom Python function of your own to every row

# Creating a function that returns the last 4 digits of a numbers
def last_four(num):
    return str(num)[-4:]

print("The Last Four Digits Test: ", last_four(123456789), "\n")

# Test passed succesfully, now let's apply it to the DF

df.head()

print("The last four numbers of the toal_bill:\n", df['total_bill'].apply(last_four))
```

**Apply With Multiple Columns**

```python
# A lambda expression is an anonymous function in programming, typically used for short, concise, and inline operations.
# We are going to create a simple function, followed by its lambda expression

# Simple function
def num(n):
    return n*2

print(num(4))

# The lambda expression of the same function
lambda num1 : num1*2
```

```python
# Applying lambda expression on a column
df['total_bill'].apply(lambda bill:bill*2)
```

**Statistical Description and Sorting**

```python
# Describing the data
df.describe()
```

```python
# We can also transpose it for readibility
df.describe().transpose()
```

```python
# Sorting the data on a column
df.sort_values('tip', ascending=True)
```

```python
# We can also sort with multiple columns
df.sort_values(['tip', 'size'])
```

```python
# We can get correlation of the data
df.corr()
```

```python
# We can get the value count of a column
df['sex'].value_counts()
```

```python
# We can get the number of unique values for a column
df['day'].nunique()
```
